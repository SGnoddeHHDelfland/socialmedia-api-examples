# Voorbeeld gebruik van PyTrends: krijg data vanuit Google Trends met Python

### Importeer de packages

```python
#import the libraries
import pandas as pd                        
from pytrends.request import TrendReq
```

### Laad trends zien

```python

pytrend = TrendReq()

#provide your search terms
kw_list=['droogte', 'warm', 'regen']

pytrend.build_payload(kw_list,
    timeframe='now 7-d', # als je naar trends.google.com gaat en een tijdsperiode selecteert kun je deze uit je URL halen
    geo='NL')

# Interest over time
regiondf = pytrend.interest_over_time()

#drop all rows that have null values in all columns
regiondf.dropna(how='all',axis=0, inplace=True)

#visualise
regiondf.plot(figsize=(20, 12), y=kw_list, ylabel='Relatieve interesse')
```
